```python
from itertools import permutations
from operator import itemgetter
```

```python
N=int(input())
count = 0
col=range(N)
sols=[]
symels = []

for guess in permutations(col):
    #generate all permutations of the board. Representing board this way means Queens cannot be placed in same row/column
    
    if N == len(set(guess[i]+i for i in col)) == len(set(guess[i]-i for i in col)):
        #set function eliminates repeat elements in a list. diagonals = guess[i] + i so if a diagonal is occupied,
        #set function will remove diagonal from guess and the len will no longer be equal to N
        
        solution = list(guess)
        sols.append(solution)
        #correct solution is found and recorded in sols.
        
        symels = transform(solution)
        mergelist = list(sols) + list(symels)
        #carry out transformation on solution and combine symmetry elements with known solutions
        
        if len(symels)+len(sols) - 1 == len(set(tuple(item) for item in mergelist)):
            #- 1 as guess solution will be repeated in sols and symels
            
            
            count += 1
            print( str(count) + ":" + "\n" +  str(tuple(symels))+"\n")
            print("\n".join(" o "*i + " Q "+" o " *(N-i-1) for i in guess)+"\n\n\n")
            
        
        
```

```python
def ver(board):
    new_board = board[::-1]
    return new_board
#vertical transformation reverses string

def hor(board):
    hor = [7-x for x in board]
    return hor


def diag(board):
    x = ver(board)
    y = hor(x)
    return y

def rot1(board):
    new_list=[]
    for x, y in enumerate(board):
        new_list.append([y, 7-x])
    new_list.sort(key=itemgetter(0))
    rotation=[]
    for i in range(len(new_list)):
        rotation.append(new_list[i][1])
    return rotation


#x,y ---> y -x
#followed by translation as rotation does not occur about origin

def rot2(board):
    new_list=[]
    for x, y in enumerate(board):
        new_list.append([7-x,7-y])
    new_list.sort(key=itemgetter(0))
    rotation=[]
    for i in range(len(new_list)):
        rotation.append(new_list[i][1])
    return rotation

#x,y ---> -x -y

def rot3(board):
    new_list=[]
    for x, y in enumerate(board):
        new_list.append([7-y, x])
    new_list.sort(key=itemgetter(0))
    rotation=[]
    for i in range(len(new_list)):
        rotation.append(new_list[i][1])
    return rotation

#x,y ---> -y x

def transform(board):
    
    transformation=[]
    
    rotboard1=rot1(board)
    rotboard2=rot2(board)
    rotboard3=rot3(board)
    

    transformation.append(rotboard1)
    transformation.append(rotboard2)
    transformation.append(rotboard3)
    
    transformation.append(hor(board))
    transformation.append(ver(board))
    
    transformation.append(hor(rotboard1))
    transformation.append(ver(rotboard1))
    
    transformation.append(hor(rotboard2))
    transformation.append(ver(rotboard2))

    transformation.append(hor(rotboard3))
    transformation.append(ver(rotboard3))
    
    #must carry out vertical and horizontal transformations on all rotations. gives repeat solutions which are eliminated
    #using set function.
    
    transformation.append(board)
    
    solution = set(tuple(item) for item in transformation)
    
   #8 symmetric board arrangements are expected
    
    
    return solution


    
```
